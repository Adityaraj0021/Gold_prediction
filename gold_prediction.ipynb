```python
def remove_outlier(col):
    sorted(col)
    Q1,Q3=np.percentile(col,[25,75])
    IQR=Q3-Q1
    lower_range= Q1-(1.5 * IQR)
    upper_range= Q3+(1.5 * IQR)
    return lower_range, upper_range
```

```python
for column in gold1[cont].columns:
    lr,ur=remove_outlier(gold1[column])
    gold1[column]=np.where(gold1[column]>ur,ur,gold1[column])
    gold1[column]=np.where(gold1[column]<lr,lr,gold1[column])
```

```python
plt.figure(figsize=(10,10))
gold1[cont].boxplot(vert=0)
plt.title('After Outlier Removal',fontsize=16)
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn import metrics

X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.25,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.transform(X_test)
```

```python
print('X_train',X_train.shape)
print('X_test',X_test.shape)
print('Y_train',Y_train.shape)
print('Y_test',Y_test.shape)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV
from sklearn import tree
```

```python
# Insiate param grid for which to search
param_grid = {'n_estimators': [50, 80, 100],'max_depth': [3, 5, 7]}

# create instance of the Randomforest regressor
rf = RandomForestRegressor()

# Define Girdsearch with random forest
# object parameter grid scoring and cv
grid_search = GridSearchCV(rf, param_grid, scoring='r2', cv=2)

# Fit the GridSearchCV object to the training data

grid_search.fit(X_train, Y_train)

# Print the best parameter values and score
print('Best parameter values: ', grid_search.best_params_)
print('Best score: ', grid_search.best_score_)
```

```python
rf = RandomForestRegressor(n_estimators=100)
```

```python
y_pred = rf.predict(X_test)
error_score = metrics.r2_score(Y_test, y_pred)
print("R squared error : ", error_score)
y_test_list = list(Y_test)
```

```python

from sklearn.metrics import mean_squared_error, mean_absolute_error
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline 
from sklearn.linear_model import Lasso
from xgboost import XGBRegressor
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error 
from sklearn.model_selection import GridSearchCV
```

```python
##lasso regression
poly = PolynomialFeatures(degree=2)
lasso = Lasso()
param_grid = {'lasso__alpha': [1e-4, 1e-3, 1e-2,1e-1, 1, 5, 10, 20, 30, 40]}
pipeline = make_pipeline(poly, lasso)
lasso_grid_search = GridSearchCV(pipeline,param_grid, scoring='r2', cv=3)
lasso_grid_search.fit(X_train, Y_train)
y_pred = lasso_grid_search.predict(X_train)
r2 = metrics.r2_score(Y_train, y_pred)
print("R-squared: ", r2)
print('Best parameter values: ',
lasso_grid_search.best_params_)
print('Best score: ',
lasso_grid_search.best_score_)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Create a DataFrame to store the grid search results
results_df = pd.DataFrame(lasso_grid_search.cv_results_)

# Extract relevant columns from the results DataFrame
results_df = results_df[['param_lasso__alpha', 'mean_test_score', 'std_test_score']]

# Sort the results by mean_test_score in descending order
results_df = results_df.sort_values(by='mean_test_score', ascending=False)

# Print the top results with their corresponding hyperparameters and scores
print("Top Lasso Regression Results:")
print(results_df.head(10))

# Visualize the results in a bar chart
plt.figure(figsize=(10, 6))
plt.barh(results_df['param_lasso__alpha'].astype(str), results_df['mean_test_score'], xerr=results_df['std_test_score'])
plt.xlabel('R-squared Score')
plt.ylabel('Lasso Alpha')
plt.title('Lasso Regression Grid Search Results')
plt.gca().invert_yaxis()  # Invert the y-axis to show the best results at the top
plt.show()
```
